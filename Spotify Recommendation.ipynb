#### Checking outlier percentage for each numeric column using IQR

```python
def calculate_outlier_percentage(df):
    outlier_percentages = {}

    for column in ['popularity', 'duration_ms', 'danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 
                   'liveness', 'valence', 'tempo']:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
        num_outliers = outliers.shape[0]
        total_values = df[column].shape[0]
        
        percent_outliers = (num_outliers / total_values) * 100
        outlier_percentages[column] = percent_outliers
    
    return outlier_percentages

outlier_percentages = calculate_outlier_percentage(df)
df_outlier_percentages = pd.DataFrame(outlier_percentages, index=[0]).transpose()
df_outlier_percentages.columns = ['value']
df_outlier_percentages.sort_values(by='value', ascending=False)
```

#### Checking what will be the percent of records retained **if** we remove all outliers

```python
o_cols = ['instrumentalness', 'speechiness', 'liveness', 'loudness', 'duration_ms', 'danceability', 'tempo', 'popularity']

df_c = df.copy()

for col in o_cols:
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = df_c[col].quantile(0.25)
    Q3 = df_c[col].quantile(0.75)
    
    # Calculate IQR
    IQR = Q3 - Q1

    # Define the lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filter out the outliers to get the cleaned data
    df_c = df_c[(df_c[col] >= lower_bound) & (df_c[col] <= upper_bound)]

print('**', df.shape)
print('**', df_c.shape)
```

#### Instead, we will correct skewness and circle back to outliers

**Why Correct Skewness?**
 - Skewed data often has larger variance. In distance-based algorithms, features with larger variances can dominate the distance calculations,
potentially overshadowing other important features
 - Correcting skewness can help in more accurately grouping similar data points together, which is crucial for k-NN where the nearest neighbors
are determined based on distance metrics
 - More balanced feature distributions can lead to improved accuracy and stability of the k-NN classifier

#### Checking skewness of numeric columns

```python
numeric_cols = ['popularity', 'duration_ms', 'danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 
                'liveness', 'valence', 'tempo']
df[numeric_cols].skew()
```

#### We will be trying various techniques for data transformation of above columns - to see which methods yields in least skewness for each field

#### We will 'shifting' values for Log, Square Root, and, BoxCox to handle negative values

#### Log Transformation

```python
import numpy as np

df_c = df.copy()

for col in numeric_cols:
    shift_value = abs(df_c[col].min()) + 1
    shifted_data = df_c[col] + shift_value
    df_c[col] = np.log(shifted_data)

df_c[numeric_cols].skew()
```

#### Square Root Transformation

```python
df_c = df.copy()

for col in numeric_cols:
    shift_value = abs(df_c[col].min()) + 1
    shifted_data = df_c[col] + shift_value
    df_c[col] = np.sqrt(shifted_data)

df_c[numeric_cols].skew()
```

#### BoxCox Transformation

```python
from scipy.stats import boxcox

df_c = df.copy()

for col in numeric_cols:
    shift_value = abs(df_c[col].min()) + 1
    shifted_data = df_c[col] + shift_value
    df_c[col], _ = boxcox(shifted_data)

df_c[numeric_cols].skew()
```

#### Yeo-Johnson Transformation

```python
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer(method='yeo-johnson')

df_c = df.copy()

df_c[numeric_cols] = pt.fit_transform(df_c[numeric_cols])

df_c[numeric_cols].skew()
```

#### We are going to use NearestNeighbors algorithm for building recommendation system

```python
from sklearn.neighbors import NearestNeighbors
```

#### We will be using 'cosine similarity' metric for finding nearest neighbors
#### - 'auto' parameter will attempt to decide the most appropriate algorithm

```python
model_knn = NearestNeighbors(n_neighbors=11, metric='cosine', algorithm='auto')
model_knn.fit(df_features)
```

#### This model will take user input track_id and return 10 nearest neighbors

```python
def recommend_tracks(track_id, df, model_knn, n_neighbors=11):
    track_index = df[df['track_id'] == track_id].index[0]
    print(track_index)
    distances,  indices = model_knn.kneighbors(df_features.iloc[track_index].values.reshape(1,-1), n_neighbors=n_neighbors)
    rec_indices = indices.flatten()[1:]
    return rec_indices
```

```python
rec_indices = recommend_tracks('5dcIjv7GMNMSaQ7nAkoFNM', df, model_knn)
rec_indices
```

```python
track_ids = []
artists = []
album_names = []
track_names = []
track_genres = []

for i in rec_indices:
    track_ids.append(df.iloc[i]['track_id'])
    artists.append(df.iloc[i]['artists'])
    album_names.append(df.iloc[i]['album_name'])
    track_names.append(df.iloc[i]['track_name'])
    track_genres.append(df.iloc[i]['track_genre'])

data_dict1 = {
    'Track ID': track_ids,
    'Artist': artists,
    'Album Name': album_names,
    'Track Name': track_names,
    'Track Genres': track_genres
}

recommend_df1 = pd.DataFrame(data_dict1)
recommend_df1
```

```python
df_arr.shape
```

```python
# inner product (dot product) index
index = faiss.IndexFlatIP(15)
```

```python
index.add(df_arr)
```

```python
k = 11  # number of nearest neighbors
D, I = index.search(df_arr, k)
```
